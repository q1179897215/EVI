```python
import numpy as np
import matplotlib.pyplot as plt

mu = 100
sigma = 15
x = mu + sigma * np.random.randn(200)
num_bins = 100

plt.figure(figsize=(12,9), dpi=200)

n, bins, patches = plt.hist(x, num_bins, color='whitesmoke', alpha=1, edgecolor='black', density=True, label='Count')

# y = ((1/(np.sqrt(2*np.pi)*sigma))*np.exp(-0.5*(1/sigma*(bins-mu))**2))
# plt.plot(bins, y, '--', label='Density', color='dimgray', linewidth=2)

# Add a vertical line at the mean
mean_val = x.mean()
label_mean = 90

plt.axvline(mean_val, color='r', linestyle='dashed', linewidth=2, label='Prediction Mean')
plt.axvline(label_mean, color='g', linestyle='dashed', linewidth=2, label='Label Mean')

# Display the mean value on the plot
plt.text(mean_val+1, 0.03, 'Mean: {:.2f}'.format(mean_val), fontsize=20, color='r')

plt.xlabel('CVR prediction values', fontsize=28, color='black')
plt.ylabel('Count', fontsize=28, color='black')
plt.xticks(fontsize=20, color='black')
plt.yticks(fontsize=20, color='black')
# Change the background color
# plt.gca().set_facecolor('lightgray')

plt.grid(axis='y', alpha=0.75)
plt.style.use('fast')
plt.legend(fontsize=20)
plt.savefig('histogram.png', dpi=200)
plt.show()
```
